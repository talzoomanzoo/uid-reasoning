### Download all data into pre-filtered directory

```python
import csv
import json
from tqdm import tqdm
from datasets import load_dataset, Dataset
import os

if not os.path.exists("./pre-filtered"):
    os.makedirs("./pre-filtered")

#datasets to download
# talzoomanzoo/0612_1.5B_dpo_zy_zthreshold
# talzoomanzoo/0612_1.5B_rft_zy_zthreshold
# talzoomanzoo/0612_1.5B_dpo_zy_ytrue
# talzoomanzoo/0612_1.5B_rft_zy_ytrue

# talzoomanzoo/0612_1.5B_dpo_zy_zthreshold
dpo_zy_zthreshold = load_dataset("talzoomanzoo/0612_1.5B_dpo_zy_zthreshold", split="train")
with open("./pre-filtered/0612_1.5B_dpo_zy_zthreshold.json", "w") as f:
    dpo_zy_zthreshold_filtered = []
    for item in tqdm(dpo_zy_zthreshold):
        dpo_zy_zthreshold_filtered.append({
            'idx': item['idx'],
            'id': item['id'],
            'input_x': item['input_x'],
            'chosen_z': item['chosen_z'],
            'chosen_y_under_chosen_z': item['chosen_y_under_chosen_z'],
            'rejected_y_under_chosen_z': item['rejected_y_under_chosen_z'],
            'rejected_z' : item['rejected_z'],
            'chosen_y_under_rejected_z': item['chosen_y_under_rejected_z'],
            'rejected_y_under_rejected_z': item['rejected_y_under_rejected_z'],
            'chosen_zy': item['chosen_z_y'],
            'rejected_zy': item['rejected_z_y'],
            'ground_truth': item['ground_truth'],
        })
    print(len(dpo_zy_zthreshold_filtered))
    json.dump(dpo_zy_zthreshold_filtered, f, ensure_ascii=False, indent=4)


# talzoomanzoo/0612_1.5B_rft_zy_zthreshold
rft_zy_zthreshold = load_dataset("talzoomanzoo/0612_1.5B_rft_zy_zthreshold", split="train")
with open("./pre-filtered/0612_1.5B_rft_zy_zthreshold.json", "w") as f:
    rft_zy_zthreshold_filtered = []
    for item in tqdm(rft_zy_zthreshold):
        rft_zy_zthreshold_filtered.append({
            'idx': item['idx'],
            'id': item['id'],
            'input_x': item['input_x'],
            'chosen_z': item['chosen_z'],
            'chosen_y_under_chosen_z': item['chosen_y_under_chosen_z'],
            'rejected_y_under_chosen_z': item['rejected_y_under_chosen_z'],
            'rejected_z' : item['rejected_z'],
            'chosen_y_under_rejected_z': item['chosen_y_under_rejected_z'],
            'rejected_y_under_rejected_z': item['rejected_y_under_rejected_z'],
            'chosen_zy': item['chosen_zy'],
            'rejected_zy': item['rejected_zy'],
            'ground_truth': item['ground_truth'],
        })
    print(len(rft_zy_zthreshold_filtered))
    json.dump(rft_zy_zthreshold_filtered, f, ensure_ascii=False, indent=4)

# talzoomanzoo/0612_1.5B_dpo_zy_ytrue
dpo_zy_ytrue = load_dataset("talzoomanzoo/0612_1.5B_dpo_zy_ytrue", split="train")
with open("./pre-filtered/0612_1.5B_dpo_zy_ytrue.json", "w") as f:
    dpo_zy_ytrue_filtered = []
    for item in tqdm(dpo_zy_ytrue):
        dpo_zy_ytrue_filtered.append({
            'idx': item['idx'],
            'id': item['id'],
            'input_x': item['input_x'],
            'chosen_z': item['chosen_z'],   
            'rejected_z' : item['rejected_z'],
            'chosen_zy': item['chosen_zy'],
            'rejected_zy': item['rejected_zy'],
            'ground_truth': item['ground_truth'],
            'metrics': item['metrics'],
        })
    print(len(dpo_zy_ytrue_filtered))
    json.dump(dpo_zy_ytrue_filtered, f, ensure_ascii=False, indent=4)

# talzoomanzoo/0612_1.5B_rft_zy_ytrue
rft_zy_ytrue = load_dataset("talzoomanzoo/0612_1.5B_rft_zy_ytrue", split="train")
with open("./pre-filtered/0612_1.5B_rft_zy_ytrue.json", "w") as f:
    rft_zy_ytrue_filtered = []
    for item in tqdm(rft_zy_ytrue):
        rft_zy_ytrue_filtered.append({
            'idx': item['idx'],
            'id': item['id'],
            'input_x': item['input_x'],
            'chosen_z': item['chosen_z'],
            'chosen_zy': item['chosen_zy'],
            'ground_truth': item['ground_truth'],
            'metrics': item['metrics'],
        })
    print(len(rft_zy_ytrue_filtered))
    json.dump(rft_zy_ytrue_filtered, f, ensure_ascii=False, indent=4)
```

```python
#dpo_zy_zthreshold filtering
import os
if not os.path.exists("./post-filtered"):
    os.makedirs("./post-filtered")
# Check and filter out rows where chosen_zy is over 22528
print("Original length of dpo_zy_zthreshold_filtered:", len(dpo_zy_zthreshold_filtered))

# Count rows where chosen_zy is over 22528
over_threshold_count = sum(1 for item in dpo_zy_zthreshold_filtered if len(item['chosen_zy']) > 22528)
print(f"Number of rows where chosen_zy > 22528: {over_threshold_count}")

# Filter out rows where chosen_zy is over 22528
dpo_zy_zthreshold_filtered = [item for item in dpo_zy_zthreshold_filtered if len(item['chosen_zy']) <= 22528]

print("Filtered length of dpo_zy_zthreshold_filtered:", len(dpo_zy_zthreshold_filtered))

# Save the filtered data back to the file
with open("./post-filtered/0612_1.5B_dpo_zy_zthreshold.json", "w") as f:
    json.dump(dpo_zy_zthreshold_filtered, f, ensure_ascii=False, indent=4)

print("Filtered data saved back to file.")
```
