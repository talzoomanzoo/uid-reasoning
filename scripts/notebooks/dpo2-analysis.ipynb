```python
import json

file_path = '/home/hojinkim/mjgwak/search-o1-dev/scripts/outputs/aime.ds-qwen-7b.direct.step-2.test/filtered_data/filtered_data2.json'

# Dictionary to store differences
different_predictions = {}

with open(file_path, 'r') as f:
    data = json.load(f)
    
    # Iterate through each of the 199 lists
    for list_index, list_data in enumerate(data):
        print(f"Checking list {list_index}...")
        
        # For each of the 100 rows in this list
        for row in list_data:
            row_id = row.get('new_id', '')
            
            # Get all prediction values for this row
            pred_values = []
            for key in row.keys():
                if key.startswith('Pred_Answer_'):
                    pred_values.append(row[key])

            print(pred_values)
            
            # If we have multiple predictions and they're different
            if len(pred_values) > 1 and len(set(pred_values)) > 1:
                # Store this case
                if list_index not in different_predictions:
                    different_predictions[list_index] = []
                
                pred_dict = {k: row[k] for k in row.keys() if k.startswith('Pred_Answer_')}
                different_predictions[list_index].append({
                    'row_id': row_id,
                    'predictions': pred_dict
                })
                # Print immediately for visibility
                print(f"\nFound different predictions in list {list_index}, row {row_id}:")
                for k, v in pred_dict.items():
                    print(f"  {k}: {v}")

# Save results
with open('different_predictions.json', 'w') as f:
    json.dump(different_predictions, f, indent=2)

# Print summary
total_lists_with_differences = len(different_predictions)
total_rows_with_differences = sum(len(rows) for rows in different_predictions.values())
print(f"\nSummary:")
print(f"Found differences in {total_lists_with_differences} lists")
print(f"Total rows with different predictions: {total_rows_with_differences}")
print("Results saved to different_predictions.json")
```

```python
import json

file_path = '/home/hojinkim/mjgwak/search-o1-dev/scripts/outputs/aime.ds-qwen-7b.direct.step-2.test/filtered_data/filtered_data.json'

# Dictionary to store differences
different_predictions = {}

with open(file_path, 'r') as f:
    data = json.load(f)
    
    # Iterate through each of the 199 lists
    for list_index, list_data in enumerate(data):
        print(f"Checking list {list_index}...")
        
        # Process in groups of 10 rows
        for group_start in range(0, len(list_data), 10):
            group_end = group_start + 10
            pred_values = []
            
            # Get predictions for this group of 10 rows
            for row in list_data[group_start:group_end]:
                for key in row.keys():
                    if key.startswith('Pred_Answer_'):
                        pred_values.append(row[key])
            
            print(pred_values)
            
            # If we have multiple predictions and they're different
            if len(pred_values) > 1 and len(set(pred_values)) > 1:
                # Store this case
                if list_index not in different_predictions:
                    different_predictions[list_index] = []
                
                group_info = {
                    'group_range': f"{group_start}-{group_end-1}",
                    'predictions': pred_values,
                    'unique_predictions': list(set(pred_values))
                }
                different_predictions[list_index].append(group_info)
                
                # Print immediately for visibility
                print(f"\nFound different predictions in list {list_index}, rows {group_start}-{group_end-1}:")
                print(f"  All values: {pred_values}")
                print(f"  Unique values: {list(set(pred_values))}")

# Save results
with open('different_predictions_by_groups.json', 'w') as f:
    json.dump(different_predictions, f, indent=2)

# Print summary
total_lists_with_differences = len(different_predictions)
total_groups_with_differences = sum(len(groups) for groups in different_predictions.values())
print(f"\nSummary:")
print(f"Found differences in {total_lists_with_differences} lists")
print(f"Total groups with different predictions: {total_groups_with_differences}")
print("Results saved to different_predictions_by_groups.json")
```

```python
import json

file_path = '/home/hojinkim/mjgwak/search-o1-dev/scripts/outputs/aime.ds-qwen-7b.direct.step-2.test/filtered_data/filtered_data.json'

# Dictionary to store differences
different_predictions = {}
lists_with_mixed_math_equal = set()  # Track lists with both True and False math_equal values

with open(file_path, 'r') as f:
    data = json.load(f)
    
    # Iterate through each of the 199 lists
    for list_index, list_data in enumerate(data):
        print(f"Checking list {list_index}...")
        
        # Process in groups of 10 rows
        for group_start in range(0, len(list_data), 10):
            group_end = group_start + 10
            pred_values = []
            math_equal_values = []
            
            # Get predictions and math_equal values for this group of 10 rows
            for row in list_data[group_start:group_end]:
                for key in row.keys():
                    if key.startswith('Pred_Answer_'):
                        pred_values.append(row[key])
                    elif key.startswith('Metrics_') and 'math_equal' in row[key]:
                        math_equal_values.append(row[key]['math_equal'])
            
            print(pred_values)
            
            # If we have multiple predictions and they're different
            if len(pred_values) > 1 and len(set(pred_values)) > 1:
                # Store this case
                if list_index not in different_predictions:
                    different_predictions[list_index] = []
                
                group_info = {
                    'group_range': f"{group_start}-{group_end-1}",
                    'predictions': pred_values,
                    'unique_predictions': list(set(pred_values)),
                    'math_equal_values': math_equal_values
                }
                different_predictions[list_index].append(group_info)
                
                # Check if this group has both True and False math_equal values
                if True in math_equal_values and False in math_equal_values:
                    lists_with_mixed_math_equal.add(list_index)
                
                # Print immediately for visibility
                print(f"\nFound different predictions in list {list_index}, rows {group_start}-{group_end-1}:")
                print(f"  All values: {pred_values}")
                print(f"  Unique values: {list(set(pred_values))}")
                print(f"  Math equal values: {math_equal_values}")

# Save results
with open('different_predictions_by_groups.json', 'w') as f:
    json.dump(different_predictions, f, indent=2)

# Print summary
total_lists_with_differences = len(different_predictions)
total_groups_with_differences = sum(len(groups) for groups in different_predictions.values())
print(f"\nSummary:")
print(f"Found differences in {total_lists_with_differences} lists")
print(f"Total groups with different predictions: {total_groups_with_differences}")
print(f"Number of lists with both True and False math_equal values: {len(lists_with_mixed_math_equal)}")
print("Results saved to different_predictions_by_groups.json")
```
