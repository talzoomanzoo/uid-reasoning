```python
#1. dpo_z_y_true (adapter)
from huggingface_hub import create_repo, upload_folder

repo_id = "talzoomanzoo/0603_dpo_z_y_true_qlora"
local_model_path = "/scratch/hojinkim/search-o1-dev/OpenRLHF/mj_scripts/checkpoint/deepseek-7b-dpo-z-ytrue"

create_repo(repo_id, exist_ok=True, private=True)

upload_folder(
    repo_id=repo_id,
    folder_path=local_model_path,
    repo_type="model",
    commit_message="0603_dpo_z_y_true_qlora"
)
```

```python
#1-1. dpo_z_y_true (merged)

from huggingface_hub import create_repo, upload_folder
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import os

# Define base and adapter
base_model_name = "deepseek-ai/DeepSeek-R1-Distill-Qwen-7B"
adapter_path = "talzoomanzoo/0603_dpo_z_y_true_qlora"

# Load base model
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_name,
    device_map="auto",
    trust_remote_code=True,
)

# Load adapter and merge
model = PeftModel.from_pretrained(base_model, adapter_path)
model = model.merge_and_unload()

# Save merged model locally
save_dir = "./0603_dpo_z_y_true_merged"
model.save_pretrained(save_dir)

tokenizer = AutoTokenizer.from_pretrained(base_model_name, trust_remote_code=True)
tokenizer.save_pretrained(save_dir)

# Upload merged model to Huggingface
repo_id = "talzoomanzoo/0603_dpo_z_y_true_merged"
create_repo(repo_id, exist_ok=True, private=False)

upload_folder(
    repo_id=repo_id,
    folder_path=save_dir,
    repo_type="model",
    commit_message="dpo_z_y_true_merged"
)
```
