```python
#sft y run3 merged
from huggingface_hub import create_repo, upload_folder
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import os

# === Define base and adapter ===
base_model_name = "deepseek-ai/DeepSeek-R1-Distill-Qwen-7B"
adapter_path = "talzoomanzoo/0507-distill-qwen-7b-sft-y-run3-qlora"

# === Load base model ===
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_name,
    device_map="auto",
    trust_remote_code=True,
)

# === Load adapter and merge ===
model = PeftModel.from_pretrained(base_model, adapter_path)
model = model.merge_and_unload()

# === Save merged model locally ===
save_dir = "./0507-DeepSeek-R1-Distill-Qwen-7B-sft-y-run3-qlora-merged"
model.save_pretrained(save_dir)

# === (Optional) Save tokenizer for completeness ===
tokenizer = AutoTokenizer.from_pretrained(base_model_name, trust_remote_code=True)
tokenizer.save_pretrained(save_dir)

# === Upload to Hugging Face ===
repo_id = "talzoomanzoo/0507-DeepSeek-R1-Distill-Qwen-7B-sft-y-run3-qlora-merged"
create_repo(repo_id, exist_ok=True, private=False)

upload_folder(
    repo_id=repo_id,
    folder_path=save_dir,
    repo_type="model",
    commit_message="Initial upload of sft-y-run3-qlora fine-tuned model"
)
```
