```python
import json

expert_dir = "../outputs/temp_06/expert/s1-32B/jama_full_test"
naive_dir = "../outputs/temp_06/naive/s1-32B/jama_full_test"

differences = []

for i in range(355):
    with open(f"{expert_dir}/output_{i}.json", "r") as f_expert:
        expert_data = json.load(f_expert)
    
    with open(f"{naive_dir}/output_{i}.json", "r") as f_naive:
        naive_data = json.load(f_naive)
    
    if expert_data["is_correct"] != naive_data["is_correct"]:
        differences.append({
            "index": i,
            "expert_is_correct": expert_data["is_correct"],
            "naive_is_correct": naive_data["is_correct"]
        })

# Save the differences to a JSON file
output_file = "./outputs/differences_s1-32B.json"
with open(output_file, "w") as f:
    json.dump(differences, f, indent=4)
```
